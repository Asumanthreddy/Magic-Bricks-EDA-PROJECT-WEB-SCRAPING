```python
import time
import pandas as pd
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# ---------------- CHROME OPTIONS ----------------
options = Options()
# Headless OFF for debugging (can turn ON later)
# options.add_argument("--headless")
options.add_argument("--window-size=1920,1080")
options.add_argument("--disable-gpu")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
    "AppleWebKit/537.36 (KHTML, like Gecko) "
    "Chrome/120.0.0.0 Safari/537.36"
)

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

data = []

# Hyderabad MagicBricks listings
for page in range(1, 6):
    url = f"https://www.magicbricks.com/property-for-sale/residential-real-estate?cityName=Hyderabad&page={page}"
    driver.get(url)

    time.sleep(6)  # wait for JS to load

    # Scroll to load cards
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(4)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    cards = soup.select("div.m-srp-card")

    print(f"Page {page} -> Listings found: {len(cards)}")

    for card in cards:
        def get_text(selector):
            tag = card.select_one(selector)
            return tag.get_text(strip=True) if tag else "N/A"

        data.append({
            "Price": get_text("div.m-srp-card__price"),
            "Location": get_text("div.m-srp-card__address"),
            "Property Type": get_text("div.m-srp-card__type"),
            "Area": get_text("div.m-srp-card__area"),
            "Posted Date": get_text("div.m-srp-card__posted-date")
        })

driver.quit()

# Save CSV
df = pd.DataFrame(data)
df.to_csv("magicbricks_hyderabad_properties.csv", index=False)

df
```
